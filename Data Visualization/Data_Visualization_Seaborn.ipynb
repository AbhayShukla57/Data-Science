# Data Visualization Seaborn

### Data: titanic.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
sns.set_style('whitegrid')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
sns.jointplot(x='Fare',y='Age',data=titanic)
```

```python
sns.displot(titanic['Fare'],kde=False,color='red',bins=30)
```

```python
sns.boxplot(x='Pclass',y='Age',data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x='Pclass',y='Age',data=titanic,palette='Set2')
```

```python
sns.countplot(x='Sex',data=titanic)
```

```python
sns.heatmap(titanic.drop(['Name','Sex','Ticket','Cabin','Embarked'],axis=1).corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='Sex')
g.map(plt.hist,'Age')
```
